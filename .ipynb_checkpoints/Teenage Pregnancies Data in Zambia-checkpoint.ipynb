```python
data = pd.read_csv('Teen pregnancies in Zambia.csv')
```

```python
data
```

```python
lua = data[data.Province == 'LUAPULA']
```

```python
lua
```

```python
lsk = data[data.Province == 'LUSAKA']
```

```python
plt.plot(lua.YearOfConception, lua.Pregnancies)
plt.plot(lsk.YearOfConception, lsk.Pregnancies)
plt.legend(['Luapula', 'Lusaka'])
plt.xlabel('Year')
plt.ylabel('No. of pregnancies')
plt.show()
```

```python
lua.Pregnancies
```

```python
lua.Pregnancies / lua.Pregnancies.iloc[0] * 100
```

```python
plt.plot(lua.YearOfConception, lua.Pregnancies / lua.Pregnancies.iloc[0] * 100)
plt.plot(lsk.YearOfConception, lsk.Pregnancies / lsk.Pregnancies.iloc[0] * 100)
plt.legend(['Luapula', 'Lusaka'])
plt.xlabel('Year')
plt.ylabel('Rise in Pregnancies(first year = 100)')
plt.show()
```
